```python
import yfinance as yf
import os
```

```python
os.makedirs("data/raw", exist_ok=True)
```

```python
tickers = {
    "gold":  "GC=F",  # gold
    "dxy":   "DX-Y.NYB", #US Dollar Index
    "vix":   "^VIX", #วัดความผันผวนคลาดหุ้น
    "yield": "^TNX", # Bond yield USA
    "sp500": "^GSPC", 
    "oil":    "CL=F", # WTI Crude Oil
    
}
```

```python
for name, ticker in tickers.items():
    df = yf.download(ticker, start="2015-01-01", end="2026-03-31")
    
    df.columns = df.columns.get_level_values(0)
    
    df.reset_index(inplace=True)
    
    df.to_csv(f"../data/raw/{name}_raw.csv", index=False)
    print(f"Saved {name}_raw.csv — {len(df)} rows")
```

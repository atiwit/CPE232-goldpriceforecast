```python
import pandas as pd
```

```python
gold = pd.read_csv('../data/raw/gold_raw.csv')
dol= pd.read_csv('../data/raw/dxy_raw.csv')
vix = pd.read_csv('../data/raw/vix_raw.csv')
yi = pd.read_csv('../data/raw/yield_raw.csv')
sp = pd.read_csv('../data/raw/sp500_raw.csv')
oil = pd.read_csv('../data/raw/oil_raw.csv')
```

```python
datasets = {
    "gold": gold,
    "dol": dol,
    "vix": vix,
    "yi": yi,
    "sp": sp,
    "oil": oil
}
```

```python
# แปลง object เป็น datetime
for name, df in datasets.items():
    datasets[name]["Date"] = pd.to_datetime(datasets[name]["Date"])
```

```python
# ตรวจสอบ
for name, df in datasets.items():
    print(f"{name}: {df['Date'].dtype}")
```

```python
#  Align + Merge
df = datasets["gold"][["Date", "Close", "High", "Low", "Open", "Volume"]].copy()
df.columns = ["Date", "gold_close", "gold_high", "gold_low", "gold_open", "gold_vol"]

secondary = {
    "dxy":   datasets["dol"],
    "vix":   datasets["vix"],
    "yield": datasets["yi"],
    "sp500": datasets["sp"],
    "oil":   datasets["oil"],
}
```

```python
for name, other_df in secondary.items():
    temp = other_df[["Date", "Close"]].rename(columns={"Close": f"{name}_close"})
    df = df.merge(temp, on="Date", how="left")

df.sort_values("Date", inplace=True)
df.ffill(inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

print("After merge:")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print(f"Missing: {df.isnull().sum().sum()}")
```

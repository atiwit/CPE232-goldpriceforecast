# Logistic Regression

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
# Load train and validation data
X_train = pd.read_csv('../../../data/processed/splits/X_train.csv').drop(columns=['Date'], errors='ignore')
y_train = pd.read_csv('../../../data/processed/splits/y_train_clf.csv')['target_direction']
X_val = pd.read_csv('../../../data/processed/splits/X_val.csv').drop(columns=['Date'], errors='ignore')
y_val = pd.read_csv('../../../data/processed/splits/y_val_clf.csv')['target_direction']
```

```python
# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
# Train Logistic Regression with cross-validation and regularization
logreg_cv = LogisticRegressionCV(
    Cs=20,
    cv=5,
    penalty='l2',
    solver='lbfgs',
    multi_class='multinomial',
    max_iter=1000,
    scoring='accuracy',
    n_jobs=-1,
    refit=True
)
logreg_cv.fit(X_train_scaled, y_train)
```

```python
# Evaluate on validation set
val_preds = logreg_cv.predict(X_val_scaled)
print('Validation Accuracy:', accuracy_score(y_val, val_preds))
print('Confusion Matrix:\n', confusion_matrix(y_val, val_preds))
print('Classification Report:\n', classification_report(y_val, val_preds))
```

```python
# Show best regularization strength
print('Best C (Inverse Regularization):', logreg_cv.C_)
```

```python
# Check validation accuracy for reference
val_preds = logreg_cv.predict(X_val_scaled)
print('Validation Accuracy:', accuracy_score(y_val, val_preds))
```

```python
# Evaluate on X_test
X_test = pd.read_csv('../../../data/processed/splits/X_test.csv').drop(columns=['Date'], errors='ignore')
y_test = pd.read_csv('../../../data/processed/splits/y_test_clf.csv')['target_direction']
X_test_scaled = scaler.transform(X_test)
test_preds = logreg_cv.predict(X_test_scaled)
print('Test Accuracy:', accuracy_score(y_test, test_preds))
print('Confusion Matrix:\n', confusion_matrix(y_test, test_preds))
print('Classification Report:\n', classification_report(y_test, test_preds))
```

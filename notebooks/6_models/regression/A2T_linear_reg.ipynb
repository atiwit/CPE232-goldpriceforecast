# Linear Regression Model

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
# Load train, validation, and test data
X_train = pd.read_csv('../../../data/processed/splits/X_train.csv').drop(columns=['Date'], errors='ignore')
y_train = pd.read_csv('../../../data/processed/splits/y_train_reg.csv')['target_return']
X_val = pd.read_csv('../../../data/processed/splits/X_val.csv').drop(columns=['Date'], errors='ignore')
y_val = pd.read_csv('../../../data/processed/splits/y_val_reg.csv')['target_return']
X_test = pd.read_csv('../../../data/processed/splits/X_test.csv').drop(columns=['Date'], errors='ignore')
y_test = pd.read_csv('../../../data/processed/splits/y_test_reg.csv')['target_return']
```

```python
# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train Ridge Regression with cross-validation
alphas = np.logspace(-4, 4, 50)
ridge_cv = RidgeCV(alphas=alphas, cv=5, scoring='neg_mean_squared_error')
ridge_cv.fit(X_train_scaled, y_train)
print('Best alpha:', ridge_cv.alpha_)
```

```python
# Evaluate on validation set
val_preds = ridge_cv.predict(X_val_scaled)
print('Validation RMSE:', mean_squared_error(y_val, val_preds, squared=False))
print('Validation MAE:', mean_absolute_error(y_val, val_preds))
print('Validation R2:', r2_score(y_val, val_preds))
```

```python
# Evaluate on test set
test_preds = ridge_cv.predict(X_test_scaled)
print('Test RMSE:', mean_squared_error(y_test, test_preds, squared=False))
print('Test MAE:', mean_absolute_error(y_test, test_preds))
print('Test R2:', r2_score(y_test, test_preds))
```
